Part 2 – Visualization of Data using matplotlib, pyplots Pack-ages

```python
import matplotlib.pyplot as plt
#data
people=['kiran','arun','vijay','varun']
age=[25,30,35,40]
height=[145,151,165,173]
weight=[45,55,65,75]

#generate graph
plt.scatter(age,height,color='blue')
plt.title('Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height(cm)')
plt.show()
```

```python
plt.scatter(age, height, color='blue')
plt.title('Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.show()
```

```python
plt.bar(people, weight, color='green')
plt.title('Weight of Individuals')
plt.xlabel('People')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
plt.hist(age, bins=5, color='red', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Explanation and Justification

Scatter Plot (Age vs Height):

Title: Age vs Height X-axis: Age Y-axis: Height (cm) Explanation: This scatter plot shows the relationship between age and height. As age increases, height also tends to increase.

Bar Chart (Weight of Individuals):

Title: Weight of Individuals X-axis: People Y-axis: Weight (kg) Explanation: This bar chart represents the weight of each individual. It provides a clear comparison of the weights, showing who is the heaviest and who is the lightest.

Histogram (Age Distribution):

Title: Age Distribution X-axis: Age Y-axis: Frequency Explanation: This histogram shows the distribution of ages in the dataset.
